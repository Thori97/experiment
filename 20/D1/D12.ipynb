```python
import csv
```

```python
with open('AC30mm/F0000CH1.CSV', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーを読み飛ばしたい時
    i=0
    x1 = []
    y1 = []
    for row in reader:
        i+=1
        if i>17:
            x1.append(float(row[3]))
            y1.append(float(row[4]))
with open('AC30mm/F0000CH2.CSV', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーを読み飛ばしたい時
    i=0
    x2 = []
    y2 = []
    for row in reader:
        i+=1
        if i>17:
            x2.append(float(row[3]))
            y2.append(float(row[4]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x1, y1, label = "CH1")
plt.plot(x2, y2, label = "CH2")
plt.legend()
plt.title("AC30mm")
plt.show()
```

```python
p = (2*0.95+1*0.03+1*0.015+0*0.005)/2
p
```

```python
import numpy as np
```

```python
-1*p*np.log2(p)-1*(1-p)*np.log2(1-p)
```

```python
-0.95*np.log2(0.95)-0.03*np.log2(0.03)-0.015*np.log2(0.015)-0.005*np.log2(0.005)
```

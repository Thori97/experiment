```python
import csv
```

```python
with open('AC30mm/F0000CH1.CSV', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーを読み飛ばしたい時
    i=0
    x1 = []
    y1 = []
    for row in reader:
        i+=1
        if i>17:
            x1.append(float(row[3]))
            y1.append(float(row[4]))
with open('AC30mm/F0000CH2.CSV', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーを読み飛ばしたい時
    i=0
    x2 = []
    y2 = []
    for row in reader:
        i+=1
        if i>17:
            x2.append(float(row[3]))
            y2.append(float(row[4])*50)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x1, y1, label = "CH1")
plt.plot(x2, y2, label = "CH2")
plt.legend()
#plt.title("AC30mm")
plt.xlabel("t[s]")
plt.savefig("D11.png", dpi = 500)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import japanize_matplotlib
```

```python
# ButterworthHPF
f = open("butterworth_HPF.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()


plt.plot(F, G)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("BGHPF.png")
plt.show()

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("位相[deg]")
plt.savefig("BPHPF.png")
plt.show()
```

```python
# ButterworthLPF
f = open("Butterworth_LPF_bode.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()

for i in range(0, len(P)):#位相が途中でひっくり返っていたので直す
    if P[i]>0:
        P[i] -= 360


plt.plot(F, G)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("BGLPF.png")
plt.show()

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("位相[deg]")
plt.savefig("BPLPF.png")
plt.show()
```

```python
#ButterHPFの方形波応答波形
f = open("Butterworth_HPF_step.txt")
for i in range(0, 6):
    line = f.readline()
T1 = []#時間1
T2 = []#時間2
W1 = []#波形1
W2 = []#波形2

while line:
    line = line.split()
    T1.append(float(line[1].split(":")[2]))
    T2.append(float(line[4].split(":")[2]))
    W1.append(float(line[2]))
    W2.append(float(line[5]))
    line = f.readline()
f.close()

w1, = plt.plot(T1, W1)
w2, = plt.plot(T2, W2)
plt.xlabel("time [s]")
plt.ylabel("Voltage [V]")
plt.title("waveform")
plt.legend([w1, w2], ["input", "output"])
plt.show()
```

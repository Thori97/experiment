```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import japanize_matplotlib
```

```python
# activeHPF
f = open("active_HPF_bode.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()


plt.plot(F, G)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("GactH.png")
plt.show()

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("振幅[dB]")
plt.savefig("PactH.png")
plt.show()
```

```python
# activeLPF
f = open("active_LPF_bode.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()

for i in range(0, len(P)):#位相が途中でひっくり返っていたので直す
    if P[i]>0:
        P[i] -= 360

plt.plot(F, G)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("GactL.png")
plt.show()

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("振幅[dB]")
plt.savefig("PactL.png")
plt.show()
```

```python
#activeHPFの方形波応答波形
f = open("active_HPF_step.txt")
for i in range(0, 6):
    line = f.readline()
T1 = []#時間1
T2 = []#時間2
W1 = []#波形1
W2 = []#波形2

while line:
    line = line.split()
    T1.append(float(line[1].split(":")[2]))
    T2.append(float(line[4].split(":")[2]))
    W1.append(float(line[2]))
    W2.append(float(line[5]))
    line = f.readline()
f.close()

w1, = plt.plot(T1, W1)
w2, = plt.plot(T2, W2)
plt.xlabel("時間[s]")
plt.ylabel("電圧[V]")
plt.legend([w1, w2], ["input", "output"])
plt.savefig("WactH.png")
plt.show()
```

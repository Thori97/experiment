```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import japanize_matplotlib
```

```python
#微分回路Cfなし
f = open("Bode_differential_withoutCf.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()
for i in range(0, len(P)):#位相が途中でひっくり返っていたので直す
    if P[i]>0:
        P[i] -= 360

plt.plot(F, G)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("dGRr.png")
plt.show()

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("位相[deg]")
plt.savefig("dPRr.png")
plt.show()
```

```python
#発振回路
f = open("Bode_hassin.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()


plt.plot(F, G)
plt.xscale("log")
plt.xlabel("Freq [Hz]")
plt.ylabel("Gain [dB]")
plt.title("Freq-Gain")
plt.show()

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("Freq [Hz]")
plt.ylabel("Phase [deg]")
plt.title("Freq-Phase")
plt.show()
```

```python
#発振回路のスペクトル
#txtの最後に余計な改行があるから消しておく
f = open("spec_hassin.txt")
for i in range(0, 409):
    line = f.readline()
F = []#周波数
S = []#スペクトル
i=0
while line:
    line = line.split()
    F.append(float(line[0]))
    S.append(float(line[1]))
    line = f.readline()
f.close()

plt.plot(F, S)
#plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("スペクトル")
plt.savefig("spha.png")
plt.show()
print(F[S.index(max(S))])
```

```python
#発振回路の応答？1512Hz
f = open("wave_hassin1512Hz.txt")
for i in range(0, 6):
    line = f.readline()
T1 = []#時間1
W1 = []#波形1

while line:
    line = line.split()
    T1.append(float(line[1].split(":")[2]))
    W1.append(float(line[2]))
    line = f.readline()
f.close()

plt.plot(T1, W1)
plt.xlabel("time [s]")
plt.ylabel("Voltage [V]")
plt.title("waveform")
plt.show()
```

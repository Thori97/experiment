```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
%matplotlib inline
import japanize_matplotlib
```

```python
#20dBの反転増幅回路
f = open("Bode20dB.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()


print(((G[15]-G[16])/np.log10(F[15]/F[16])))
print(G[0])
print(G[16])
print(P[0])
print(F[16])
print(F)
print(G)
print(P)

plt.plot(F, G)
plt.xscale("log")
plt.xlabel("周波数 [Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("G20dB.png")
plt.show()

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("周波数 [Hz]")
plt.ylabel("位相[deg]")
plt.savefig("P20dB.png")
plt.show()
```

```python
#微分回路完全体
f = open("Bode_differentialcircuit.txt")
for i in range(0, 4):
    line = f.readline()
F = []#周波数
G = []#ゲイン
P = []#位相

while line:
    line = line.split()
    F.append(float(line[0]))
    G.append(float(line[1]))
    P.append(float(line[2]))
    line = f.readline()
f.close()


print(G)
plt.plot(F, G)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("dGRrCf.png")
plt.show()

for i in range(0, len(P)):#位相が途中でひっくり返っていたので直す
    if P[i]>0:
        P[i] -= 360

plt.plot(F, P)
plt.xscale("log")
plt.xlabel("周波数[Hz]")
plt.ylabel("ゲイン[dB]")
plt.savefig("dPRrCf.png")
plt.show()
```

```python
#20dB非反転増幅回路の応答波形(1kHz)
f = open("wave20dB_notreverse_1kHz.txt")
for i in range(0, 6):
    line = f.readline()
T1 = []#時間1
T2 = []#時間2
W1 = []#波形1
W2 = []#波形2

while line:
    line = line.split()
    T1.append(float(line[1].split(":")[2]))
    T2.append(float(line[4].split(":")[2]))
    W1.append(float(line[2]))
    W2.append(float(line[5]))
    line = f.readline()
f.close()
print(max(W1))
print(max(W2))


w1, = plt.plot(T1, W1)
w2, = plt.plot(T2, W2)
plt.xlabel("時間[s]")
plt.ylabel("電圧[V]")
plt.savefig("waven20dB.png")
plt.legend([w1, w2], ["input", "output"])
plt.show()
```

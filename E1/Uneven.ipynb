```python
import matplotlib.pyplot as plt
import numpy as np

from pylab import rcParams

rcParams['figure.figsize'] = 12,8
plt.rcParams["font.size"] = 18

import scipy as sp
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return m, h
```

```python
# 球-平板
ball_board_DC_plus_raw = np.array([
    # d [mm], Vs[kV]
    np.array([5, 17.85, 17.86, 17.97, 17.85, 17.83]),
    np.array([10, 32.44, 32.51, 32.42, 32.49, 32.54]),
    np.array([15, 45.16, 45.34, 45.53, 44.91, 45.31]),
    np.array([20, 56.06, 56.01, 55.73, 55.76, 55.92])
])

ball_board_DC_plus = np.array([
    [d[0], np.average(d[1:]), 3 * np.std(d[1:])] for d in ball_board_DC_plus_raw
])

print(ball_board_DC_plus)
```

```python
ball_board_DC_minus_raw = np.array([
    # d [mm], Vs[kV]
    np.array([5, 17.57, 17.93, 17.84, 17.89, 17.85]),
    np.array([10, 32.03, 32.04, 32.05, 32.09, 32.00]),
    np.array([15, 44.26, 44.59, 44.42, 44.29, 44.47]),
    np.array([20, 54.99, 55.05, 54.95, 54.88, 55.05])
])

ball_board_DC_minus = np.array([
    [d[0], np.average(d[1:]), 3 * np.std(d[1:])] for d in ball_board_DC_minus_raw
])

print(ball_board_DC_minus)
```

```python
ball_board_AC_raw = np.array([
    # d [mm], Vs[kV]
    np.array([5, 16.87, 16.84, 16.98, 16.59, 16.93]),
    np.array([10, 30.34, 30.29, 30.27, 29.92, 30.15]),
    np.array([15, 42.09, 42.10, 42.04, 42.19, 41.89]),
    np.array([20, 52.00, 51.89, 51.84, 51.88, 52.11])
])

ball_board_AC = np.array([
    [d[0], np.average(d[1:]), 3 * np.std(d[1:])] for d in ball_board_AC_raw
])

print(ball_board_AC)
```

```python
plt.title("ball-board")
plt.xlabel("$d\ [\mathrm{mm}]$")
plt.ylabel("$V_s\ [\mathrm{kV}]$")

plt.plot(ball_board_DC_plus[:, 0], ball_board_DC_plus[:, 1], label="DC plus", marker='o')
plt.plot(ball_board_DC_minus[:, 0], ball_board_DC_minus[:, 1], label="DC minus", marker='o')
plt.plot(ball_board_AC[:, 0], ball_board_AC[:, 1], label="AC", marker='o')

#plt.errorbar(ball_board_DC_plus[:, 0], ball_board_DC_plus[:, 1], yerr = ball_board_DC_plus[:, 2], fmt='o', label="DC plus")
#plt.errorbar(ball_board_DC_minus[:, 0], ball_board_DC_minus[:, 1], yerr = ball_board_DC_minus[:, 2], fmt='o', label="DC minus")
#plt.errorbar(ball_board_AC[:, 0], ball_board_AC[:, 1], yerr = ball_board_AC[:, 2], fmt='o', label="AC")

plt.legend()
```

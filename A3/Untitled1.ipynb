```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib
```

```python
data = [
    #[n, t_pLH, t_pHL]
    [2, 22.4e-9, 13.6e-9],
    [1, 19.2e-9, 10.8e-9],
    [3, 45.6e-9, 14.8e-9],
    [4, 54.0e-9, 14.8e-9]
]
data  = [
    [1,2,3,4],
    [19.2e-9, 22.4e-9, 45.6e-9, 54.0e-9],
    [10.8e-9, 13.6e-9, 14.8e-9, 20.2e-9]
]
```

```python
plt.plot(data[0], data[1])
plt.show()
```

```python
n = np.array(data[0])
tL = np.array(data[1])
tH = np.array(data[2])
a1 = np.cov(n, tL)[0][1]/np.var(n)
print(a1)
b1 = np.mean(tL) - a1 * np.mean(n)
print(b1)
```

```python
plt.plot(data[0], data[1], label = '実験値')
x = np.linspace(1, 4)
plt.plot(x, a1*x+b1, label = '理論値')
plt.legend()
plt.savefig("up.jpg")
plt.show()
```

```python
n = np.array(data[0])
tL = np.array(data[1])
tH = np.array(data[2])
a2 = np.cov(n, tH)[0][1]/np.var(n)
print(a2)
b2 = np.mean(tH) - a2 * np.mean(n)
print(b2)
plt.plot(data[0], data[2], label = '実験値')
x = np.linspace(1, 4)
plt.plot(x, a2*x+b2, label = '理論値')
plt.legend()
plt.savefig("down.jpg")
plt.show()
```

```python
n = np.array(data[0])
tL = np.array(data[1])
tH = np.array(data[2])
t = (tL+tH)/2
a3 = np.cov(n, t)[0][1]/np.var(n)
print(a3)
b3 = np.mean(t) - a3 * np.mean(n)
print(b3)
plt.plot(n, t, label = '実験値')
x = np.linspace(1, 4)
plt.plot(x, a3*x+b3, label = '理論値')
plt.legend()
plt.show()
```
